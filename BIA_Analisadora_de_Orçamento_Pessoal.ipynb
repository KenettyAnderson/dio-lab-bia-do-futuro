```python
# ==============================================================================
# PROJETO: BIA - Analisadora de Orçamento Pessoal (ETL com GenAI)
# ==============================================================================

# Instalação das dependências necessárias (Descomente a linha abaixo se for rodar localmente ou no Colab)
# !pip install openai pandas -q

import os
import pandas as pd
from openai import OpenAI
from IPython.display import display

# --- CONFIGURAÇÃO DA API ---
# Substitua 'SUA_API_KEY_AQUI' pela sua chave real da OpenAI
os.environ['OPENAI_API_KEY'] = 'SUA_API_KEY_AQUI'
client = OpenAI(api_key=os.environ.get('OPENAI_API_KEY'))

print("Iniciando o processo ETL com GenAI...\n")

# ==============================================================================
# PASSO 1: EXTRAÇÃO (Extract)
# Criando os dados diretamente no código para garantir que o projeto seja
# 100% autossuficiente e à prova de erros de caminhos de arquivos (FileNotFound).
# ==============================================================================
print("⏳ [1/3] Extraindo dados do cliente...")

dados_extrato = {
    'Data': ['01/03', '05/03', '12/03', '15/03', '20/03', '25/03'],
    'Categoria': ['Supermercado', 'Delivery', 'Transporte', 'Delivery', 'Lazer', 'Delivery'],
    'Valor_R$': [450.00, 85.50, 60.00, 110.00, 200.00, 65.00]
}

df_extrato = pd.DataFrame(dados_extrato)
total_gasto = df_extrato['Valor_R$'].sum()

# ==============================================================================
# PASSO 2: TRANSFORMAÇÃO (Transform)
# Usando a IA Generativa para analisar os dados numéricos e gerar insights.
# ==============================================================================
print("🧠 [2/3] Transformando dados em inteligência com a BIA (GPT-3.5)...")

# Converte o DataFrame para uma string formatada para a IA ler
extrato_em_texto = df_extrato.to_string(index=False)

# Define o comportamento da assistente
system_prompt = """
Você é a BIA, uma assistente virtual financeira inteligente, empática e direta, baseada na assistente do Bradesco.
Sua missão é ajudar os clientes a terem uma vida financeira mais saudável analisando seus extratos.
Ao receber os dados de gastos:
1. Faça um resumo acolhedor da situação.
2. Identifique o principal ponto de atenção (onde o cliente está gastando mais, ou gastos repetitivos).
3. Dê uma dica prática, realista e educacional para economizar no próximo mês.
Seja concisa (máximo de 3 parágrafos) e utilize emojis para deixar a leitura agradável.
"""

mensagem_usuario = f"Olá BIA, gastei um total de R$ {total_gasto} este mês. Aqui está o meu extrato detalhado:\n\n{extrato_em_texto}\n\nPor favor, analise para mim."

# Faz a requisição para a API da OpenAI
resposta = client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=[
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": mensagem_usuario}
    ],
    temperature=0.7
)

relatorio_bia = resposta.choices[0].message.content

# ==============================================================================
# PASSO 3: CARREGAMENTO (Load)
# Exibindo os dados brutos e o relatório final processado.
# ==============================================================================
print("✨ [3/3] Carregamento concluído!\n")
print("=" * 50)
print("📊 EXTRATO BRUTO (Dados do Pandas):")
display(df_extrato)
print(f"💰 Total Gasto: R$ {total_gasto:.2f}")
print("=" * 50)

print("\n🤖 RELATÓRIO DA BIA:")
print("-" * 50)
print(relatorio_bia)
print("-" * 50)
```
